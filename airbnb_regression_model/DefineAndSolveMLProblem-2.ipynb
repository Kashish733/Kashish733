```python
df['price_2'] = stats.mstats.winsorize(df['price'], limits=[0.01, 0.01])
df.head()
```

```python
diff = pd.unique(df['price']-df['price_2'])
print(diff)
df = df.drop('price', axis =1)
```

```python
df_null = df.isnull().head()
print(df_null)
nan_count = np.sum(df.isnull(), axis =0)
print(nan_count)
condition = nan_count !=0
col_name = nan_count[condition].index.tolist()
print(col_name)
df[col_name].dtypes
```

```python
df['host_response_rate_2'] = df['host_response_rate'].isnull()
df['host_acceptance_rate_2'] = df['host_acceptance_rate'].isnull()
df['bedrooms_2']= df['bedrooms'].isnull()
df['beds_2']= df['beds'].isnull()
df.head()
```

```python
mean_host_response_rate = df['host_response_rate'].mean()
mean_host_acceptance_rate = df['host_acceptance_rate'].mean()
mean_bedrooms = df['bedrooms'].mean()
mean_beds = df['beds'].mean()
df['host_response_rate'].fillna(value = mean_host_response_rate, inplace = True)
df['host_acceptance_rate'].fillna(value = mean_host_acceptance_rate, inplace = True)
df['bedrooms'].fillna(value = mean_bedrooms, inplace = True)
df['beds'].fillna(value = mean_beds, inplace = True)
```

```python
for col in col_name:
    print("{} missing values count :{}".format(col, np.sum(df[col].isnull(), axis = 0)))
```

```python
colu = df.select_dtypes(exclude = ['int64', 'float64']).columns
unique_sum = {}
for col in colu:
    unique_sum[col] = df[col].nunique()
unique_sum    
```

```python
df_room_type=pd.get_dummies(df['room_type'])
df = pd.concat([df, df_room_type], axis=1)                
df_neighbourhood_group_cleansed=pd.get_dummies(df['neighbourhood_group_cleansed'])
df = pd.concat([df, df_neighbourhood_group_cleansed], axis=1)
```

```python
df = df.drop('room_type', axis = 1)
df = df.drop('neighbourhood_group_cleansed', axis =1 )
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = round(numeric_df.corr(),5)
corr_matrix
```

```python
corrs = corr_matrix['price_2']
corrs
```

```python
corrs_sorted = corrs.sort_values(ascending = False)
corrs_sorted
```

```python
print(df['host_is_superhost'].unique())
print(df['host_identity_verified'].unique())
print(df['host_has_profile_pic'].unique())
exclude = ['host_is_superhost' , 'host_identity_verified', 'host_has_profile_pic','name', 'description', 'neighborhood_overview','host_name', 'host_location', 'host_about','amenities','review_scores_value','review_scores_checkin','review_scores_communication','reviews_per_month','review_scores_rating','review_scores_cleanliness','review_scores_location']
df = df.drop(exclude, axis = 1)
```

```python
df.columns.tolist()
```

```python
df['log_price'] = np.log1p(df['price_2'])
df = df.drop('price_2', axis =1 )
```

```python
X = df.drop('log_price', axis =1)
y = df['log_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state = 1234)
```

```python
print('Begin RF Implementation...')

rf_model = RandomForestRegressor(max_depth = 32, n_estimators = 300)
rf_model.fit(X_train, y_train)

print('End')
```

```python
y_rf_pred = rf_model.predict(X_test)
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared = False)
rf_r2 = r2_score(y_test, y_rf_pred)

print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))                 
```

```python
print('Running cross validation...')
model = RandomForestRegressor(max_depth = 32, n_estimators = 300)
neg_rmse_score = cross_val_score(model, X, y, scoring = 'neg_mean_squared_error' , cv=5)
rmse_score = np.sqrt(-neg_rmse_score)
print("Cross-validated RMSE score:", rmse_score)
print("Mean RMSE:", rmse_score.mean())
print('Done')
```

```python
param_grid = {'n_estimators': [100, 200,300],'max_depth': [None, 10, 20,30]}

grid_search = GridSearchCV(RandomForestRegressor(random_state= 42),param_grid, cv=5,scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Best Params:", grid_search.best_params_)
print("Test RMSE:", rmse)
print("Test R²:", r2)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.plot([y_pred.min(), y_pred.max()], [y_test.min(), y_test.max()], 'r--') 
plt.show()
```
